## Funtion remove_parentheses_content for removing explaination after answer "Yes", which is in circle bracket ()

```python
import re

def remove_parentheses_content(dataframe, column_name):
    dataframe[column_name] = dataframe[column_name].str.replace(r'\([^()]*\)', '', regex=True)
    return dataframe
df_disability_cleaned = remove_parentheses_content(df_disability, 'Do you consider yourself to be a disabled person?')
#df_disability_cleaned
```

```python
plot_frequency_distribution_11(df_disability_cleaned, "Do you consider yourself to be a disabled person?", "Disability")
```

## Function remove_text_after_yes to make plot simplier

```python
import re

def remove_text_after_yes(dataframe, column_name):
    dataframe[column_name] = dataframe[column_name].str.replace(r'Yes.*', 'Yes', regex=True)
    return dataframe

df_disability_cleaned1 = remove_text_after_yes(df_disability_cleaned, 'Do you consider yourself to be a disabled person?')
```

```python
plot_frequency_distribution_11(df_disability_cleaned1, 'Do you consider yourself to be a disabled person?', 'Disability - Simplier')
```
